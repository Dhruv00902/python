```python
#libraries for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(x=df["Age"],color="Brown")
#hist=histogram
```

```python
plt.hist(x=df["Age"],color="red")
plt.title("Distribution of age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(x=df["Age"])
plt.title("Distribution of age")
plt.xlabel("Age")
#plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(df["Age"])
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Sex Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="bar")
sns.boxplot(df["Sex"])
plt.title("Distribution of Sex")
plt.xlabel("Sex")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="barh",color="pink")
plt.title("Sex Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
df["Pclass"].value_counts().plot(kind="bar",color="teal")
plt.title("Pcalss Distribution")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```python
emb=df["Embarked"].value_counts()
keys=emb.keys().to_list()
counts=emb.to_list()
plt.pie(x=counts,labels=keys,autopct='%1.1f%%')
```
